```python
from ae.dataset import HiddenManifold, load_mnist, load_mnist_as_dataset
import matplotlib.pyplot as plt
```

## MNIST

```python
mnist = load_mnist_as_dataset()

fig, axes = plt.subplots(2, 5, figsize=(10, 4))
for i, ax in enumerate(axes.flat):
    digit, label = mnist[i]
    digit = digit.reshape(28, 28)
    ax.imshow(digit, cmap='gray')
    ax.set_title(f'Label: {label}')
    ax.axis('off')

plt.tight_layout()
plt.show()
```

```python
data, _ = load_mnist()
print(data.shape)
```

```python
# Extract the pixel intensities of the first N digits
N = 200
pixel_intensities = data[:N].flatten()
plt.hist(pixel_intensities, bins=256, range=(-1, 1), density=True)
plt.xlabel('Pixel Intensity')
plt.ylabel('Frequency')
plt.yscale('log')
plt.title('Histogram of Pixel Intensities')
plt.show()
```

```python
# compute fraction of non-zero entries
sparsity = (data > -1).sum() / (len(data) * 784)
print(f'Sparsity: {sparsity:.2f}')
```

```python
# show a kind of "average digit" by averaging the pixel intensities of all digits
avg_digit = data.mean(dim=0)
avg_digit = avg_digit.reshape(28, 28)
plt.imshow(avg_digit, cmap='gray')
plt.axis('off')
plt.show()
```
